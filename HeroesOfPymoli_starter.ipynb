```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

#format to disply dollar values correctly
pd.options.display.float_format = '${:,.2f}'.format

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)

#loc to reference later
player_demographics = purchase_data_df.loc[:, ['Gender','SN', 'Age']]

purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
#find unique players using screen name, eliminate duplicates
unique_sn = purchase_data_df["SN"].unique()
unique_sn

total = len(unique_sn)

#create a dataframe to display results
player_count = pd.DataFrame({'Total Players': [total]})
player_count
```

## Purchasing Analysis (Total)

```python
#use basic functions to calculate total items available, the average price of items and total revenue
unique_items_count = purchase_data_df["Item ID"].value_counts()
average_purchase_price = purchase_data_df["Price"].mean()
total_purchases_count = purchase_data_df["Purchase ID"].value_counts()
revenue_total = purchase_data_df["Price"].sum()

#create summary df
purchasing_analysis_total = pd.DataFrame({'Number of Unique Items': [len(unique_items_count)],
 'Average Price': [average_purchase_price],
 'Number of Purchases': [len(total_purchases_count)],
 'Total Revenue': [revenue_total]})

purchasing_analysis_total
```

## Gender Demographics

```python
#use .loc to use columns necessary for gender counts and percents
player_demographics = purchase_data_df.loc[:, ["Gender", "SN", "Age"]]

#eliminate duplicates for accurate count of total gender players
player_demographics = player_demographics.drop_duplicates() 
gender_total_counts = player_demographics["Gender"].value_counts()
gender_demographics_percents = 100. * gender_total_counts / len(player_demographics)

#format to disply percent values correctly
pd.options.display.float_format = '{:,.2f}%'.format

#create summary df
gender_demographics = pd.DataFrame({"Total Count": gender_total_counts, "Percentage of Players": gender_demographics_percents})

gender_demographics
```


## Purchasing Analysis (Gender)

```python
#use goupby to calculate basic math based off of gender
purchase_count = purchase_data_df.groupby(['Gender']).count()['Price']
total_purchase = purchase_data_df.groupby(['Gender']).sum()['Price']
average_purchase_price = total_purchase / purchase_count
avg_purchase_total = total_purchase / gender_total_counts

#format to disply dollar values correctly
pd.options.display.float_format = '${:,.2f}'.format

#create summary df
purchasing_analysis = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": average_purchase_price, "Total Purchase Price": total_purchase, "Avg Total Purchase per Person": avg_purchase_total})
purchasing_analysis
```

## Age Demographics

```python
#use binning to combine ages into groups
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
player_demographics["Age Group"] = pd.cut(x = player_demographics['Age'], bins = bins, labels = group_names, include_lowest = False)

#drop duplicates to find total and percent based on accurate number of players
player_demographics = player_demographics.drop_duplicates()
age_range = player_demographics.groupby(['Age Group']).count()['Age']
total_age = age_range / len(player_demographics) * 100

#format to disply percent values correctly
pd.options.display.float_format = '{:,.2f}%'.format

#create summary df
age_demographics = pd.DataFrame({"Total Count": age_range, "Percentage of Players": total_age})
age_demographics
```

## Purchasing Analysis (Age)

```python
#use binning to create age groups
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
purchase_data_df["Age Group"] = pd.cut(x = purchase_data_df['Age'], bins = bins, labels = group_names, include_lowest = False)

#calculate basic math that will be based off of age groups
purchase_count = purchase_data_df.groupby(['Age Group']).count()['Age']
total_purchase_value = purchase_data_df.groupby(['Age Group']).sum()['Price']
avg_purchase_price = total_purchase_value / purchase_count

#format to disply dollar values correctly
pd.options.display.float_format = '${:,.2f}'.format

#refer back to player demographics in order to drop dupplicates for accurate per person avg
player_demographics = player_demographics.drop_duplicates()
age_range = player_demographics.groupby(['Age Group']).count()['Age']
avg_per_person = total_purchase_value / age_range

#create summary df
purchasing_analysis_age = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": avg_purchase_price, "Total Purchase Value": total_purchase_value, "Avg Total Purchase per Person": avg_per_person})
purchasing_analysis_age
```

## Top Spenders

```python
#analyze purchases by screen name/player
purchase_count_sn = purchase_data_df.groupby(['SN']).count()['Price']
total_purchase_sn = purchase_data_df.groupby(['SN']).sum()['Price']
avg_purchase_price_sn = total_purchase_sn / purchase_count_sn

#create summary df
top_spenders = pd.DataFrame({"Purchase Count": purchase_count_sn, "Average Purchase Price": avg_purchase_price_sn, "Total Purchase Value": total_purchase_sn})
top_spenders.sort_values(by='Total Purchase Value', ascending=False)
```

## Most Popular Items

```python
#create .loc to reference items and price info. 
popular_items = purchase_data_df.loc[:, ['Item ID','Item Name', 'Price']]

#use grouby with 'items' to run basic calculations on item purchases
pop_items_total = purchase_data_df.groupby(['Item ID', 'Item Name']).count()['Price']
pop_items_sum = purchase_data_df.groupby(['Item ID', 'Item Name']).sum()['Price']
item_price = pop_items_sum / pop_items_total

#create summary table
most_popular_items = pd.DataFrame({"Purchase Count": pop_items_total, "Item Price": item_price, "Total Purchase Value": pop_items_sum})

#sort df by purchase count, descending
most_popular_items.sort_values(by='Purchase Count', ascending=False)
```
